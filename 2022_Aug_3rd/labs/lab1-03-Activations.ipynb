# 활성함수의 종류 및 케라스 모델

```python
import numpy as np                
import matplotlib.pyplot as plt  
```

```python
def step(x):
    return np.array(x > 0, dtype=np.int)
```

```python
# -5.0부터 5.0까지 0.1 간격 생성
x = np.arange(-5.0, 5.0, 0.1)          
y = step(x)
```

```python
plt.title('Step Function')
plt.plot(x,y)
plt.show()
```

```python
# 시그모이드 함수 그래프를 그리는 코드
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
```

```python
y = sigmoid(x)
```

```python
plt.plot(x, y)
plt.title('Sigmoid Function')
plt.show()
```

```python
y = np.tanh(x)
```

```python
plt.plot(x, y)
plt.title('Tanh Function')
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
y = relu(x)
```

```python
plt.plot(x, y)
plt.title('Relu Function')
plt.show()
```

```python
def leaky_relu(x):
    return np.maximum(a*x, x)
```

```python
a = 0.1
y = leaky_relu(x)
```

```python
plt.plot(x, y)
plt.title('Leaky ReLU Function')
plt.show()
```

```python
y = np.exp(x) / np.sum(np.exp(x))
```

```python
plt.plot(x, y)
plt.title('Softmax Function')
plt.show()
```
